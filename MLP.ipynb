```python
from model import Model
import numpy as np
from my_dataset import MNIST, IRIS
from sklearn.model_selection import train_test_split
import common
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

# Iris Data

```python
data = IRIS.Iris()
X, y = data['data'], data['target']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
y_train = common.one_hot_encode(y_train)
y_test = common.one_hot_encode(y_test)
```

```python
print(X.shape, x_test.shape, x_train.shape)
```

#### Batch Learning

```python
ep = 1000
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.1)
mlp1 = Model()
mlp1.add_layer(4, 6, act_fn='sigmoid')
mlp1.add_layer(6, 6, act_fn='sigmoid')
mlp1.add_layer(6, 3, act_fn='softmax')
losses_batch, accs_batch = mlp1.fit(x_train, y_train, k_fold=5, validate_x=x_valid, validate_y=y_valid,
                       epochs=ep, batch_size=x_train.shape[0], loss='squarederror', learning_rate=0.001)
```

```python
pred_batch = mlp1.predict(x_test)
pred_batch = common.convert_to_binary(pred_batch)
test_batch = common.convert_to_binary(y_test)
```

```python
epoc = [i for i in range(ep)]
plt.plot(epoc,losses_batch)
plt.title("Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
epoc = [i for i in range(ep)]
plt.plot(epoc,accs_batch)
plt.title("Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
#Generate the confusion matrix
cf_matrix = confusion_matrix(test_batch, pred_batch)
print(cf_matrix)

ax = sns.heatmap(cf_matrix, annot=True, cmap='BuPu')

ax.set_title('Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
print('Performance Report (Accuracy score on test set) :', format(metrics.accuracy_score(test_batch, pred_batch), '.2f'),'\n\n')
target=[str(i) for i in range(3)]
print('Classification Report'.center(65,'-'),'\n', metrics.classification_report(test_batch, pred_batch,target_names=target))
```

# MNIST Data

```python
(x_train, y_train), (x_test, y_test) = MNIST.Mnist()
```

```python
x_train = x_train.reshape(x_train.shape[0], -1) / 255
y_train = common.one_hot_encode(y_train)
x_test = x_test.reshape(x_test.shape[0], -1) / 255
y_test = common.one_hot_encode(y_test)
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.1)
```

#### Batch Learning

```python
ep = 50
mlp4 = Model()
mlp4.add_layer(784, 200, act_fn='sigmoid')
mlp4.add_layer(200, 128, act_fn='sigmoid')
mlp4.add_layer(128, 64, act_fn='sigmoid')
mlp4.add_layer(64, 10, act_fn='softmax')
losses_mb, accs_mb = mlp4.fit(x_train, y_train, validate_x=x_valid, validate_y=y_valid,
                       epochs=ep, batch_size=x_train.shape[0], loss='squarederror', learning_rate=0.001)
```

```python
pred_mb = mlp4.predict(x_test)
pred_mb = common.convert_to_binary(pred_mb)
test_mb = common.convert_to_binary(y_test)
```

```python
epoc = [i for i in range(ep)]
plt.plot(epoc,losses_mb)
plt.title("Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
epoc = [i for i in range(ep)]
plt.plot(epoc,accs_mb)
plt.title("Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
#Generate the confusion matrix
cf_matrix = confusion_matrix(test_mb, pred_mb)
print(cf_matrix)

plt.figure(figsize = (20,15))
ax = sns.heatmap(cf_matrix, annot=True, cmap='BuPu')

ax.set_title('Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
print('Performance Report (Accuracy score on test set) :', format(metrics.accuracy_score(test_mb, pred_mb), '.2f'),'\n\n')
target=[str(i) for i in range(10)]
print('Classification Report'.center(65,'-'),'\n', metrics.classification_report(test_mb, pred_mb,target_names=target))
```
